```python
from datasets import load_dataset

ds_main = load_dataset("ar9av/charts_train_test_val")
```

```python
from tqdm import tqdm
import pandas as pd
import json
from PIL import Image

# Sample function to simulate the transformation of the "data" column
def transform_data(data):
    output_data = [
        {
            "from": "human",
            "value": f"<image>\n{data[0]['content'][0]['text']}"
        },
        {
            "from": "gpt",
            "value": data[0]['content'][1]['text']
        }
    ]
    return output_data

# Assuming `ds` is the loaded Hugging Face dataset
# Example loading: ds = load_dataset('your_dataset_name')['split_name']

# Initialize lists to store the data
image_paths = []
conversations = []

for k in ['train']:
    ds = ds_main[k]
    for i in tqdm(range(len(ds))):
        row = ds[i]

        # Save the image and store the path
        image = row['image']
        image_path = f"images/image_{i}.png"
        image.save(image_path)  # Save the image to a file
        image_paths.append(image_path)

        # Transform the data column
        data = row['data']
        conversation = transform_data(data)
        conversations.append(conversation)

    # Create a DataFrame
    df = pd.DataFrame({
        'image': image_paths,
        'conversations': conversations
    })
```

```python
df['id'] = list(range(len(df)))
```

```python
with open('data.json', 'w') as f:    
    json.dump(df[['image', 'conversations', 'id']].to_dict(orient='records'), f)
```
